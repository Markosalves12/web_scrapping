```python
import requests
```

```python
url = "https://www.gov.br/mec/pt-br/assuntos/noticias"
```

```python
response = requests.get(url)
```

```python
# print(response.content)
```

```python
from bs4 import BeautifulSoup
```

```python
noticias  = response.content
```

```python
site = BeautifulSoup(noticias, 'html.parser')
```

```python
print(site.prettify)
```

```python
noticias_feed = site.find('ul', attrs={"class":"noticias listagem-noticias-com-foto"})
```

```python
print(noticias_feed)
```

```python
print(noticias_feed.prettify)
```

```python
type(noticias_feed)
```

```python
listagem_noticias = noticias_feed.find_all('div', attrs={'class': "conteudo"})
```

```python
listagem_noticias
```

```python
titulos = list()
desc = list()
links = list()
```

```python
import pandas as pd
```

```python
len(listagem_noticias)
```

```python
for news in listagem_noticias:
    link = news.find('a')
    tittle_news = news.find('h2', attrs={'class': 'titulo'})
    descricao = news.find('span', attrs={'class': 'descricao'})
        
    if tittle_news:
        titulos.append(str(tittle_news.text).replace('\n','').strip())
        
    if descricao:
        desc.append(str(descricao.text).replace('\n','').strip())
    
    if link:
        links.append(str(link['href']).replace('\n','').strip())
    else:
        links.append('None')
```

```python
print(len(links), len(titulos), len(desc))
```

```python
df = pd.DataFrame(
    {
        'Titulos': titulos,
        'descricao': desc,
        'links': links,
    }
)
```
